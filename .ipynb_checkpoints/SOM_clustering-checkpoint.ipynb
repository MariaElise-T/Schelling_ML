## Schelling

```python
os.chdir("/Users/maria/Desktop/Schelling_ML")
train_inputs = pd.read_csv('train_inputs_hap_0.csv')  
train_outputs = pd.read_csv('train_outputs_hap_0.csv')  
train_inputs2 = pd.read_csv('train_inputs_homo_0.csv')  
for i in range(1,100):
    filepath_in = 'train_inputs_hap_'+str(i)+'.csv'
    filepath_out = 'train_outputs_hap_'+str(i)+'.csv'
    filepath_in2 = 'train_inputs_homo_'+str(i)+'.csv'
    file_in = pd.read_csv(filepath_in)
    file_out = pd.read_csv(filepath_out)
    file_in2 = pd.read_csv(filepath_in2)
    train_inputs = pd.concat([train_inputs, file_in])
    train_outputs = pd.concat([train_outputs, file_out])
    train_inputs2 = pd.concat([train_inputs2, file_in2])
```

```python
test = train_inputs.values.tolist()
test
```

```python
rows = 3
cols = 3
structure = type_conn.grid_four
network = dtwsom.DtwSom(rows, cols, structure)
```

```python
%%time
network.train(test, 20)
```

```python
n_neurons = network._size
fig, axs = plt.subplots(3,3,figsize=(20, 10), sharey=True)
for neuron_index in range(n_neurons):
    col = math.floor(neuron_index/3)
    row = neuron_index % 3
    neuron_weights = network._weights[neuron_index]
    axs[row, col].plot(np.arange(len(neuron_weights)), neuron_weights, label=str(neuron_index))
    axs[row, col].set_ylabel("Neuron: "+str(neuron_index))
plt.show()
```

```python
file1 = open('train_outputs_burnedout_0.csv', 'r')
Lines = file1.readlines()
train_outputs_burnedout = list()

for line in Lines:
    train_outputs_burnedout.append(np.fromstring(line.strip(), dtype=int, sep=','))
    
for i in range(1,100):
    filepath_in = 'train_outputs_burnedout_'+str(i)+'.csv'
    file1 = open(filepath_in, 'r')
    Lines = file1.readlines()

    for line in Lines:
        train_outputs_burnedout.append(np.fromstring(line.strip(), dtype=int, sep=','))
        
len(train_outputs_burnedout)
```

```python
rows = 3
cols = 3
structure = type_conn.grid_four
network = dtwsom.DtwSom(rows, cols, structure)
```

```python
%%time
network.train(train_outputs_burnedout, 30)
```

```python
n_neurons = network._size
fig, axs = plt.subplots(3,3,figsize=(20, 10), sharey=True)
for neuron_index in range(n_neurons):
    col = math.floor(neuron_index/3)
    row = neuron_index % 3
    neuron_weights = network._weights[neuron_index]
    axs[row, col].plot(np.arange(len(neuron_weights)), neuron_weights, label=str(neuron_index))
    axs[row, col].set_ylabel("Neuron: "+str(neuron_index))
plt.show()
```
